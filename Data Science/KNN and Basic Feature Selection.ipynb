```python

# this cell follows ths steps from the tutorials
# to do Part A of Assignment # 2

# import the packages and libraries to create the tables/dataframes
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
# create the iris dataframe
# I choose a more meaningful variable names
# Part A
iris_df = pd.read_csv('https://raw.githubusercontent.com/mpourhoma/CS4661/master/iris.csv')
#iris_df
# Get the Feature column names and create a feature matrix
iris_feature_cols = ['sepal_length','sepal_width', 'petal_length', 'petal_width']
# Iris Feature Matrix
IFM = iris_df [iris_feature_cols]
#IFM
# create a label array/vector
LV = iris_df ['species']
#LV
```

```python

# this cell follows the steps from the tutorials 
# to do part C of Assignment # 2

# part C

k = 3
iris_knn_k3 = KNeighborsClassifier(n_neighbors=k)
# import the function train_test_split from sklearn.model_selection to split the data

# train the model
iris_knn_k3.fit(IFM_training,L_training)
# Test the trained model with the testing set.

L_predict = iris_knn_k3.predict(IFM_testing)

# import the function accuracy_score from sklean.metrics
# it performs element-to-element compairson
# and returns the correct predictions

from sklearn.metrics import accuracy_score
accuracy_rate = accuracy_score(L_testing, L_predict)
# create a KNN object with k = 3


print (accuracy_rate)
```

```python

# part D

# I choose to import the libraries and packages again
# in the following cells incase I forget to
# educe errors and to make debugging easier.
# this does the calculation of the accuracy rate of each k value 
# up to 60 and store them in a dataframe

import numpy as np
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score


iris_df = pd.read_csv('https://raw.githubusercontent.com/mpourhoma/CS4661/master/iris.csv')

iris_feature_cols = ['sepal_length','sepal_width', 'petal_length', 'petal_width']

IFM = iris_df [iris_feature_cols]

LV = iris_df ['species']

IFM_training, IFM_testing, L_training, L_testing = train_test_split(IFM, LV, test_size= 0.4, random_state= 6)

k_values = []

accuracy_rates = []

# to test each k value from 1 to 60
# I ran through each k value to be sure 
# which k gives the highest accuracy 
for i in range(1, 61, 1):

    # create a new KNN model with new k
    iris_knn_ki = KNeighborsClassifier(n_neighbors=i)

    # train the new model
    iris_knn_ki.fit(IFM_training,L_training)

    # do the prediction with the trained model
    L_predict = iris_knn_ki.predict(IFM_testing)

    # use the function to get the accuracy score/rate
    accuracy_rate = accuracy_score(L_testing, L_predict)
    
    k_values.append(i)
    
    accuracy_rates.append(accuracy_rate)
    

# I choose to store the results in a data frame

result_df = pd.DataFrame()

result_df ['k value'] = k_values

result_df ['accuracy rate'] = accuracy_rates

```

```python


# part E 

# testing each single feature with the
# same steps from part B and C for the knn model
# with k = 3.
# the results are stored in a dataframe

import numpy as np
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score

iris_df = pd.read_csv('https://raw.githubusercontent.com/mpourhoma/CS4661/master/iris.csv')

iris_feature_cols = ['sepal_length','sepal_width', 'petal_length', 'petal_width']

LV = iris_df ['species']

k_value = 3

iris_knn_k3 = KNeighborsClassifier(n_neighbors=k_value)

features_list = []

ft_accuracy_rates = []

for i in range(0, 4, 1) :
    # create new iris feature matrix with current single feature
    IFM = iris_df[iris_feature_cols[i]]

    # split the data again with current iris feature matrix 
    IFM_training, IFM_testing, L_training, L_testing = train_test_split(IFM, LV, test_size= 0.4, random_state= 6)

    # the compiler seems to say this fixes the erros I was getting
    # about single data with single features
    IFM_training = IFM_training.values.reshape(-1,1)
    IFM_testing = IFM_testing.values.reshape(-1,1)
    
    # train the model again
    iris_knn_k3.fit(IFM_training,L_training)

    # predict the labels on the testing data
    L_predict = iris_knn_k3.predict(IFM_testing)

    # use the function to get the accuracy rate
    accuracy_rate = accuracy_score(L_testing, L_predict)

    #feature_list.append(iris_feature_cols[i])
    features_list.append(iris_feature_cols[i])
    ft_accuracy_rates.append(accuracy_rate)

ft_result_df = pd.DataFrame()

ft_result_df ['feature'] = features_list

ft_result_df ['accuracy rate'] = ft_accuracy_rates
```

```python


# Part F

# this cell finds the accuracy of each pair of features
# the results are saved in a dataframe

import numpy as np
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score

list_of_pairs = [ (0, 1) , (0,2), (0,3), (1,2), (1,3), (2, 3)]

#for x in list_of_pairs:

iris_df = pd.read_csv('https://raw.githubusercontent.com/mpourhoma/CS4661/master/iris.csv')

iris_feature_cols = ['sepal_length','sepal_width', 'petal_length', 'petal_width']

LV = iris_df ['species']

k_value = 3

iris_knn_k3 = KNeighborsClassifier(n_neighbors=k_value)

first_features_list = []
second_features_list = []

ft_pairs_accuracy_rates = []

for pair in list_of_pairs:
    
    current_feature_pair = [iris_feature_cols[pair[0]], iris_feature_cols[pair[1]]]

    IFM = iris_df[current_feature_pair]

    IFM_training, IFM_testing, L_training, L_testing = train_test_split(IFM, LV, test_size= 0.4, random_state= 6)

    iris_knn_k3.fit(IFM_training,L_training)

    L_predict = iris_knn_k3.predict(IFM_testing)

    first_features_list.append(iris_feature_cols[pair[0]])
    second_features_list.append(iris_feature_cols[pair[1]])


    accuracy_rate = accuracy_score(L_testing, L_predict)

    ft_pairs_accuracy_rates.append(accuracy_rate)

ft_pairs_result_df = pd.DataFrame()

ft_pairs_result_df['first feature'] = first_features_list

ft_pairs_result_df['second feature'] = second_features_list

ft_pairs_result_df['accuracy rate'] = ft_pairs_accuracy_rates

```
