```python


import numpy as np

import pandas as pd


heart_d_df = pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Heart_s.csv")


heart_d_df_transformed = pd.get_dummies (heart_d_df, prefix_sep='', prefix ='', 
                                         columns= ['Gender', 'ChestPain', 'Thal'])


heart_d_df_trans_cols = list(heart_d_df_transformed.columns)


heart_d_df_trans_cols.remove('AHD')


HDTFM = heart_d_df_transformed[heart_d_df_trans_cols]

LABEL_V = heart_d_df_transformed['AHD']



from sklearn.model_selection import train_test_split

HDTFM_training, HDTFM_testing, Label_training, Label_testing = train_test_split(HDTFM, LABEL_V, test_size = 0.25, random_state = 6)


from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score


k_value = 3

rs_tree_value = 5


k3_knn_classifier = KNeighborsClassifier(n_neighbors=k_value)

rs5_decision_tree = DecisionTreeClassifier(random_state = rs_tree_value)

max_iter_value = 1067 # min value on this computer to not reach an error
rs_log_regression = LogisticRegression(max_iter=1500)


k3_knn_classifier.fit(HDTFM_training,Label_training)

rs5_decision_tree.fit(HDTFM_training,Label_training)

rs_log_regression.fit(HDTFM_training,Label_training)


label_predict_knn = k3_knn_classifier.predict(HDTFM_testing)

label_predict_dec_tree = rs5_decision_tree.predict(HDTFM_testing)

label_predict_log_reg = rs_log_regression.predict(HDTFM_testing) 


acc_rate_knn_k3 = accuracy_score (Label_testing,label_predict_knn)

acc_rate_dec_tree = accuracy_score (Label_testing,label_predict_dec_tree)

acc_rate_log_red = accuracy_score (Label_testing,label_predict_log_reg)



print( "knn rate: " +  str(acc_rate_knn_k3))

print( "decision tree rate: " + str(acc_rate_dec_tree)) 

print ("logistic regression rate: " + str(acc_rate_log_red))


```


# Part G Results

Decision Tree accuracy result improveed quite a bit.

# Part H



```python


import numpy as np

import pandas as pd


heart_d_df = pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Heart_s.csv")

heart_d_df_transformed = pd.get_dummies (heart_d_df, prefix_sep='', prefix ='', 
                                         columns= ['Gender', 'ChestPain', 'Thal'])

heart_d_df_trans_cols = list(heart_d_df_transformed.columns)

heart_d_df_trans_cols.remove('AHD')


HDTFM = heart_d_df_transformed[heart_d_df_trans_cols]

LABEL_V = heart_d_df_transformed['AHD']



from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.neighbors import KNeighborsClassifier



k_value = 3

rs_tree_value = 5


k3_knn_classifier = KNeighborsClassifier(n_neighbors=k_value)

rs5_decision_tree = DecisionTreeClassifier(random_state = rs_tree_value)

max_iter_value = 1067 # min value on this computer to not reach an error
rs_log_regression = LogisticRegression(max_iter=1500)


from sklearn.model_selection import cross_val_score

k_folds = 10

k3_knn_acc_list = cross_val_score(k3_knn_classifier, HDTFM, LABEL_V, cv = k_folds, scoring ='accuracy')

rs5_dec_tree_acc_list = cross_val_score(rs5_decision_tree, HDTFM, LABEL_V, cv = k_folds, scoring ='accuracy')

rs_log_reg_acc_list = cross_val_score(rs_log_regression, HDTFM, LABEL_V, cv = k_folds, scoring ='accuracy')


acc_cv_k3_knn = k3_knn_acc_list.mean()

acc_cv_dec_tree = rs5_dec_tree_acc_list.mean()

acc_cv_log_reg = rs_log_reg_acc_list.mean()

print ('knn cross validation score: ' + str ( acc_cv_k3_knn) )

print ('decision tree cross validation score: ' + str(acc_cv_dec_tree) )

print ('logistic regression cross validation score: ' + str(acc_cv_log_reg)  )

```
