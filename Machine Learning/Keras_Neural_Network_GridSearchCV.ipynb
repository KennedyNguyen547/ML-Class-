
# This notebook was done as an assignment for a machine learning class. This notebook contains codes for using keras neural networks and grid search with cross validation. The dataset was images of 28x28 greyscale single digits 0-9 doing digit recognition.
# The code was compiled on AMD FX 8320 with 8GB Ram on Windows 7.



# Part One
Many of the code lines were copied from Lecture 11


```python



# "Sequential" models let us define a stack of neural network layers
from keras.models import Sequential

# import the core layers:
from keras.layers import Dense, Dropout, Activation, Flatten

import numpy as np



from  keras.wrappers.scikit_learn  import  KerasClassifier

from  sklearn.model_selection  import  GridSearchCV



# import some utilities to transform/preprocess our data:

from keras.utils import np_utils

```


# Part 2

Many of the code lines were copied from Lecture 11


```python

# Keras will download MNIST digit dataset for us:
from keras.datasets import mnist

# By default, the first 60k of MNIST has been defined as training and the rest as testing set: 
(X_train, y_train), (X_test, y_test) = mnist.load_data()




# Reshape each image pixels into a row of feature table with 28*28=784 features (each pixel is a feature):
X_train = X_train.reshape(X_train.shape[0], 784)

X_test = X_test.reshape(X_test.shape[0], 784)


X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```


# Part 3

Copied from Lecture 11


```python


y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)

```


# Part 4
Many of following code lines were taken from Lecture 11<br>
to define the fuction that creates models.


```python


def model_creator():
    
    model = Sequential()
    
    input_size = 784
    hidden_neurons = 100
    out_size = 10
    
    # -----------------------------------------
    # second layer: hidden layer:
    model.add(Dense(hidden_neurons, input_dim = input_size))  # Nuerons
    model.add(Activation('sigmoid')) # Activation

    # -----------------------------------------
    # third layer: output layer:
    model.add(Dense(out_size, input_dim = hidden_neurons))  # Nuerons
    model.add(Activation('softmax')) # Activation

    model.compile(loss='categorical_crossentropy', optimizer='adam',  metrics=['accuracy'])
    
    return model
```



# Part 5
This line was put here as following the instructions in order.


```python

seed = 2
np.random.seed(seed)

```



# Part 6




```python


model = KerasClassifier(build_fn = model_creator, verbose=2)
```



# Part 7
See Results


```python

import time

param_grid = { 'batch_size':[30 , 50 , 100 ], 'epochs':[10 , 15 , 20] }

start_time = time.perf_counter()

grid = GridSearchCV(model, param_grid, cv = 10, scoring='neg_log_loss' )

grid.fit ( X_train, y_train )

end_time = time.perf_counter()



print(grid.best_score_)
print(grid.best_params_)
```
